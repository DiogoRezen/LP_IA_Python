```python
import json
import matplotlib.pyplot as plt
import numpy as np

'''
stack_file_name_E = []
for i in range(21):
    stack_file_name_E.append('stack_file'+str(i+1)+'_E')
'''
def mean_std(vetor):
  media = np.average(vetor)
  desvi = np.std(vetor)

  return media, desvi

def z_score(vetor):
  media, std = mean_std(vetor)
  z_filtro = []
  for i in range(len(vetor)):
    z_filtro.append((vetor[i] - media) / std)
  
  return z_filtro

def normal(vetor):
  minimo = np.min(vetor)
  maximo = np.max(vetor)

  z = []
  z1 = []
  for i in range(len(vetor)):
    z.append( (vetor[i] - minimo) / (maximo - minimo) )
  
  for i in range(len(vetor)):
    if z[i] > 0:
      z1.append(z[i])

  return z

temp_file_s1e = []; stack_file_s1e = []
temp_file_s2e = []; stack_file_s2e = []
temp_file_s3e = []; stack_file_s3e = []
temp_file_s4e = []; stack_file_s4e = []
temp_file_s5e = []; stack_file_s5e = []
temp_file_s6e = []; stack_file_s6e = []
temp_file_s7e = []; stack_file_s7e = []
temp_file_s8e = []; stack_file_s8e = []
temp_file_s9e = []; stack_file_s9e = []

for i in range(21):
    file_name = 'dados'+str(i+1)+'.json'
    with open(file_name,'r') as openfile:
        json_object = json.load(openfile)
    temp_file_s1e.append(json_object['s1e'])
    temp_file_s2e.append(json_object['s2e'])
    temp_file_s3e.append(json_object['s3e'])
    temp_file_s4e.append(json_object['s4e'])
    temp_file_s5e.append(json_object['s5e'])
    temp_file_s6e.append(json_object['s6e'])
    temp_file_s7e.append(json_object['s7e'])
    temp_file_s8e.append(json_object['s8e'])
    temp_file_s9e.append(json_object['s9e'])
    #print(len(temp_file_s1e[i]))

for i in range(len(temp_file_s1e)):
    for j in range(len(temp_file_s1e[i])):
        stack_file_s1e.append(temp_file_s1e[i][j])
        stack_file_s2e.append(temp_file_s2e[i][j])
        stack_file_s3e.append(temp_file_s3e[i][j])
        stack_file_s4e.append(temp_file_s4e[i][j])
        stack_file_s5e.append(temp_file_s5e[i][j])
        stack_file_s6e.append(temp_file_s6e[i][j])
        stack_file_s7e.append(temp_file_s7e[i][j])
        stack_file_s8e.append(temp_file_s8e[i][j])
        stack_file_s9e.append(temp_file_s9e[i][j])
        #print(temp_file_s1e[i][j])

x = [i for i in range(len(stack_file_s1e))]

print(len(stack_file_s1e),len(x))
#medias1e,stds1e = mean_std(stack_file_s1e)
#print(medias1e)

s1e_filtro_z = z_score(stack_file_s1e); s1e_normal = normal(stack_file_s1e)
s2e_filtro_z = z_score(stack_file_s2e); s2e_normal = normal(stack_file_s2e)
s3e_filtro_z = z_score(stack_file_s3e); s3e_normal = normal(stack_file_s3e)

s6e_filtro_z = z_score(stack_file_s6e); s6e_normal = normal(stack_file_s6e)
s7e_filtro_z = z_score(stack_file_s7e); s7e_normal = normal(stack_file_s7e)
s8e_filtro_z = z_score(stack_file_s8e); s8e_normal = normal(stack_file_s8e)

s9e_filtro_z = z_score(stack_file_s9e); s9e_normal = normal(stack_file_s9e)

plt.figure(figsize = ((12,6)))
plt.plot(s1e_filtro_z[0:100])

xlim = 50
plt.figure(figsize = ((24,6)))
plt.plot(x[0:xlim],s1e_normal[0:xlim])
plt.plot(x[0:xlim],s2e_normal[0:xlim])
plt.plot(x[0:xlim],s3e_normal[0:xlim])
plt.legend(loc="upper left")
medias1eF,stds1eF = mean_std(s1e_normal)
print(medias1eF)

plt.figure(figsize = ((24,6)))
plt.plot(x[0:xlim],s6e_normal[0:xlim])
plt.plot(x[0:xlim],s7e_normal[0:xlim])
plt.plot(x[0:xlim],s8e_normal[0:xlim])
plt.legend(loc="upper left")
#plt.figure(figsize = ((12,6)))
#plt.subplot(1,3,1)
#plt.plot(stack_file_s1e[0:100])

plt.figure(figsize = ((24,6)))
plt.plot(x[0:xlim],s9e_normal[0:xlim])



       

```
